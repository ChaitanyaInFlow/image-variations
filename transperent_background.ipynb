```python
def change_background(input_path, background_path):
    img = Image.open(input_path).convert('RGB') # read image
    fig, ax = plt.subplots(1,2,figsize=(8,4))
    ax[0].imshow(img)
    ax[0].axis('off')
#     out = remover.process(img) # default setting - transparent background
#     out = remover.process(img, type='rgba')
    remover = Remover() # default setting
    out = remover.process(img, type=background_path) # use another image as a background
    ax[1].imshow(out)
    ax[1].axis('off')
    plt.show()
```
